```python
#Bài tập for 2: Game21
import random
print("This is 21 game!")
while True:
    r = random.Random()
    current_number = 0
    current_player = ""
    player_turn = r.randint(0,1)
    if player_turn == 0:
        current_player = "human"
        print("Player go first!")
    else:
        current_player = "computer"
        print("Computer go first!")

    while current_number <= 21:
        if current_player == "human":
            player_choice = int(input("Player's turn. Input number between 1 and 3:"))
            while player_choice == 1 or player_choice == 2 or player_choice == 3:
                current_number += player_choice
                print(f"Current number is {current_number}")
                if current_number > 21:
                    print("Current number is greater than 21. You lose!")
                    break
                else:
                    current_player = "computer"
                break
            else:
                print("You must input number between 1 and 3. Please input again!")    
        if current_player == "computer": 
            computer_choice = r.randint(1,3)
            print(f"Computer's turn. Computer's choice is {computer_choice}")
            current_number += computer_choice
            print(f"Current number is {current_number}")
            if current_number > 21:
                print("Current number is greater than 21. You win!")
                break
            else:
                current_player = "human"
    if input("Do you want to play again? (Y/N)") != "y":
        break
```
